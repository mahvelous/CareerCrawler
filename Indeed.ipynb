```python
from selenium import webdriver
import pandas as pd
```

```python
url = "https://www.indeed.co.in/jobs?q=machine+learning&l=Hyderabad"
driver = webdriver.Chrome()
driver.get(url)
print("executed")
```

```python
jobs =  driver.find_elements_by_class_name('jobsearch-SerpJobCard')
print(len(jobs))
```

```python
job_list=[]
i=0
for job in  jobs:
    try:
        cname = job.find_element_by_xpath('div[1]/div[1]/span').text
    except:
        cname = "Not Available"
    try:
        position = job.find_element_by_xpath('h2/a').get_attribute('title')
    except:
        position = "Not Available"
    try:
        website = job.find_element_by_xpath('h2/a').get_attribute('href')
    except:
        website = "Not Available"
    try:
        link = job.find_element_by_xpath('h2/a').get_attribute('href')
    except:
        link = "Not Available"
    try:
        location = job.find_element_by_xpath('div[1]/div[3]').text
    except:
        location = "Not Available"
    try:
        experience = job.find_element_by_xpath('').text
    except:
        experience ="Not Available"
    try:
        salary = job.find_element_by_xpath('div[2]/span/span').text
    except:
        salary ="Not Available"
    job_item={'Company Name':cname,
          'Position':position,
          'Location':location,
          'Salary':salary,
          'Company Website':website,
          'Apply link':link
         }
    job_list.append(job_item)
    i+=1
    if i==10:
        break
df = pd.DataFrame(job_list)
df
```
