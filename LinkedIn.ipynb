```python
from selenium import webdriver
import pandas as pd
url = "https://in.linkedin.com/jobs/search?keywords=Machine%20Learning&location=Hyderabad"
driver = webdriver.Chrome()
driver.get(url)
print("executed")
```

```python
job_list=[]
jobs =  driver.find_elements_by_class_name('result-card__contents')
print(len(jobs))
```

```python
for job in  jobs:
    try:
        cname = job.find_element_by_xpath('h4/a').text
    except:
        cname = "Not Available"
    try:
        position = job.find_element_by_xpath('h3').text
    except:
        position = "Not Available"
    try:
        website =job.find_element_by_xpath('h4/a').get_attribute('href')
    except:
        website = "Not Available"
    try:
        link = job.find_element_by_xpath('h4/a').get_attribute('href')
    except:
        link = "Not Available"
    try:
        location = job.find_element_by_xpath('div/span[1]').text
    except:
        location = "Not Available"
    job_item={'Company Name':cname,
          'Position':position,
          'Location':location,
          'Company Website':website,
          'Apply link':link,
         }
    job_list.append(job_item)
df = pd.DataFrame(job_list)
df
```
