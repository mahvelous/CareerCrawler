```python
from selenium import webdriver
import pandas as pd
print("Enter profile")
profile = input()
print(profile)
print("Enter location")
location=input()
print(location)
print("Experience")
experience=input()
print(experience)
```

```python
url = "https://www.monsterindia.com/srp/results?limit=10&query="+profile+"&locations="+location+"&experienceRanges="+experience+"~"+experience+"&experience="+experience
driver = webdriver.Chrome()
driver.get(url)
print("executed")
```

```python
job_list=[]
jobs =  driver.find_elements_by_class_name('card-apply-content')
for job in  jobs:
    try:
        cname = job.find_element_by_xpath('div/span/a').text
    except:
        cname = "Not Available"
    try:
        position = job.find_element_by_xpath('div/h3/a').text
    except:
        position = "Not Available"
    try:
        website =job.find_element_by_xpath('div/span/a').get_attribute('href')
    except:
        website = "Not Available"
    try:
        link = job.find_element_by_xpath('div/h3/a').get_attribute('href')
    except:
        link = "Not Available"
    try:
        location = job.find_element_by_xpath('div/div/div[1]/span/small').text
    except:
        location = "Not Available"
    try:
        experience = job.find_element_by_xpath('div/div/div[2]/span/small').text
    except:
        experience ="Not Available"
    try:
        salary = job.find_element_by_xpath('div/div/div[3]/span/small').text
    except:
        salary = "Not Available"
    try:
        description = job.find_element_by_xpath('p[1]').text
    except:
        description = "Not Available"
    job_item={'Company Name':cname,
          'Position':position,
          'Location':location,
          'Experience':experience,
          'Salary':salary,
          'Company Website':website,
          'Apply link':link,
          'Description': description
         }
    job_list.append(job_item)
df = pd.DataFrame(job_list)
df
```
