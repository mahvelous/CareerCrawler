```python
from selenium import webdriver
import pandas as pd
```

```python
print("Enter profile")
profile = input()
print(profile)
print("Enter location")
location=input()
print(location)
print("Experience")
experience=input()
print(experience)
```

```python
url = "https://www.naukri.com/"+profile+"-jobs-in-"+location
driver = webdriver.Chrome()
driver.get(url)
print("executed")
```

```python
job_list=[]
jobs =  driver.find_elements_by_class_name('jobTuple')
print(len(jobs))
```

```python
job_list=[]
i=0
for job in  jobs:
    try:
        cname = job.find_element_by_xpath('div[1]/div[1]/div/a').text
    except:
        cname = "Not Available"
    try:
        position = job.find_element_by_xpath('div[1]/div[1]/a').get_attribute('title')
    except:
        position = "Not Available"
    try:
        website = job.find_element_by_xpath('div[1]/div[1]/div/a').get_attribute('href')
    except:
        website = "Not Available"
    try:
        link = job.find_element_by_xpath('div[1]/div[1]/a').get_attribute('href')
    except:
        link = "Not Available"
    try:
        location = job.find_element_by_xpath('div[1]/div[1]/ul/li[3]/span').text
    except:
        location = "Not Available"
    try:
        experience = job.find_element_by_xpath('div[1]/div[1]/ul/li[1]/span').text
    except:
        experience ="Not Available"
    try:
        salary = job.find_element_by_xpath('div[1]/div[1]/ul/li[2]/span').text
    except:
        salary = "Not Available"
    job_item={'Company Name':cname,
          'Position':position,
          'Location':location,
          'Experience':experience,
          'Company Website':website,
          'Apply link':link,
          'Salary':salary
         }
    job_list.append(job_item)
    i+=1
    if i==10:
        break
df = pd.DataFrame(job_list)
df
```
