```python
profile = profile.replace(" ","-")
print(profile)
url = "https://www.shine.com/job-search/machine-learning-java-jobs-in-bangalore?fexp=3"
driver = webdriver.Chrome()
driver.get(url)
print("executed")
```

```python
job_list=[]
jobs =  driver.find_elements_by_class_name('search_listing')
print(len(jobs))
```

```python
job_list=[]
i=0
for job in  jobs:
    try:
        cname = job.find_element_by_xpath('ul/li[2]/a/ul/li[3]').text
    except:
        cname = "Not Available"
    try:
        position = job.find_element_by_xpath('ul/li[2]/a').get_attribute('title')
    except:
        position = "Not Available"
    try:
        website = job.find_element_by_xpath('ul/li[2]/a').get_attribute('href')
    except:
        website = "Not Available"
    try:
        link = job.find_element_by_xpath('ul/li[2]/a').get_attribute('href')
    except:
        link = "Not Available"
    try:
        location = job.find_element_by_xpath('ul/li[2]/a/ul/li[4]/em').text
    except:
        location = "Not Available"
    try:
        experience = job.find_element_by_xpath('ul/li[2]/a/ul/li[4]/span').text
    except:
        experience ="Not Available"
    job_item={'Company Name':cname,
          'Position':position,
          'Location':location,
          'Experience':experience,
          'Company Website':website,
          'Apply link':link
         }
    job_list.append(job_item)
    i+=1
    if i==10:
        break
df = pd.DataFrame(job_list)
df
```
