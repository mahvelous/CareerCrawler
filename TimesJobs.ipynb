```python
from selenium import webdriver
import pandas as pd
print("Enter profile")
profile = input()
print(profile)
print("Enter location")
location=input()
print(location)
print("Experience")
experience=input()
print(experience)
```

```python
url = "https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords="+profile+"&txtLocation="+location+"&cboWorkExp1="+experience
driver = webdriver.Chrome()
driver.get(url)
print("executed")
```

```python
job_list=[]
print(job_list)
joblist = driver.find_element_by_class_name('new-joblist')
jobs = joblist.find_elements_by_tag_name('li')
len(jobs)
```

```python
job_list=[]
i=0
while i<50:
    try:
        cname = jobs[i].find_element_by_xpath('header/h3').text
    except:
        cname = "Not Available"
    try:
        position = jobs[i].find_element_by_xpath('header/h2/a').text
    except:
        position = "Not Available"
    website = "Not Available"
    try:
        link = jobs[i].find_element_by_xpath('header/h2/a').get_attribute('href')
    except:
        link = "Not Available"
    try:
        location = jobs[i].find_element_by_xpath('ul[1]/li[2]/span').get_attribute('title')
    except:
        location = "Not Available"
    try:
        experience = jobs[i].find_element_by_xpath('ul[1]/li[1]').text
    except:
        experience ="Not Available"
    salary = "Not Available"
    try:
        description = jobs[i].find_element_by_xpath('ul[2]/li[1]').text
    except:
        description = "Not Available"
    job_item={'Company Name':cname,
          'Position':position,
          'Location':location,
          'Experience':experience[12:],
          'Salary':salary,
          'Company Website':website,
          'Apply link':link,
          'Description': description
         }
    if cname!="Not Available"
        job_list.append(job_item)
    i+=5
df = pd.DataFrame(job_list)
df
```
